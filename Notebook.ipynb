```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("Desktop/Aurangabad_Transport.csv")
df.head()
df.info()
df.isnull().sum()
df["Public Transport - Average No. of daily trips"] = pd.to_numeric(df["Public Transport - Average No. of daily trips"], errors="coerce")
df["Public Transport - Average No. of passengers (day-to-day basis)"] = pd.to_numeric(df["Public Transport - Average No. of passengers (day-to-day basis)"],errors="coerce")
df["Estimated_Daily_Revenue"] = (
    df["Public Transport - Revenue per km"] *
    df["Public Transport - Average Trip length in km (daily)"] *
    df["Public Transport - Average No. of daily trips"]
)

df["Estimated_Daily_Cost"] = (
    df["Public Transport - Average cost per km (daily)"] *
    df["Public Transport - Average Trip length in km (daily)"] *
    df["Public Transport - Average No. of daily trips"]
)

df["Estimated_Daily_Profit"] = df["Estimated_Daily_Revenue"] - df["Estimated_Daily_Cost"]

df["Passengers_per_Trip"] = (
    df["Public Transport - Average No. of passengers (day-to-day basis)"] /
    df["Public Transport - Average No. of daily trips"]
)
plt.figure()
plt.plot(df["Year"], df["Estimated_Daily_Revenue"])
plt.plot(df["Year"], df["Estimated_Daily_Cost"])
plt.title("Revenue vs Cost Trend")
plt.xlabel("Year")
plt.ylabel("Amount")
plt.show()
plt.figure()
plt.bar(df["Year"], df["Estimated_Daily_Profit"])
plt.title("Daily Profit by Year")
plt.xlabel("Year")
plt.ylabel("Profit")
plt.show()
```
